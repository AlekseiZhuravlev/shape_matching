```python
import numpy as np

def target_func(x_i, a):
    return a * x_i * (1 - x_i)


def evaluate_func(x_0, a):
    x = x_0
    
    x_list = []
    for i in range(100):
        x = target_func(x, a)
        x_list.append(x)
        
    return x_list
```

```python
import matplotlib.pyplot as plt

a_list = [3.3, 3.51, 3.75]
x_0_list = [0.228734167, 0.228734168]

for a in a_list:
    fig, ax = plt.subplots(1, 3, figsize=(15, 5))
    x_0_0 = evaluate_func(x_0_list[0], a)
    x_0_1 = evaluate_func(x_0_list[1], a)
    
    print('x_0_0:', x_0_0)
    print('x_0_1:', x_0_1)
    
    ax[0].plot(x_0_0)
    ax[0].set_title(f"x_0={x_0_list[0]}")
    ax[0].set_xlabel("n")
    ax[0].set_ylabel("x_n")
    
    ax[1].plot(x_0_1)
    ax[1].set_title(f"x_0={x_0_list[1]}")
    ax[1].set_xlabel("n")
    ax[1].set_ylabel("x_n")
    
    difference = np.array(x_0_0) - np.array(x_0_1)
    ax[2].plot(difference)
    ax[2].set_title(f"Difference")
    # ax[2].set_xlabel("n")
    ax[2].set_ylabel("Difference")
    
    plt.suptitle(f"a={a}")
    plt.show()

```
